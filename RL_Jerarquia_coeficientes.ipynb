```python
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

### EJERCICIO 1: MODELO SUPERVISADO.

```python
df = pd.read_excel('/Users/acrhonoz./Desktop/base_modelo_supervisado.xlsx')
```

```python
df
```

```python
# Variables predictoras
X_multiple = df.iloc[:,1:9]
```

```python
X_multiple
```

```python
#Etiquetas 
Y_multiple = df.iloc[:,9:10]
```

```python
Y_multiple
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_multiple,Y_multiple,test_size=0.2)
```

```python
rl_multiple = linear_model.LinearRegression()
```

```python
rl_multiple.fit(x_train,y_train)
```

```python
print('DATOS DEL MODELO DE REGRESIÃ“N LINEAL MULTIPLE')
print()
print('valor de los coeficientes: ')
print()
print(rl_multiple.coef_)
print()
print('valor de la intersecciÃ³n:')
print()
print(rl_multiple.intercept_)
```

```python
#PrecisiÃ³n del Modelo
print('PrecisiÃ³n del modelo: ')
print(rl_multiple.score(x_train,y_train))
```

```python
# Importancia de las caracterÃ­sticas
importances = pd.DataFrame(data={
    'Attribute': x_train.columns,
    'Importance': rl_multiple.coef_[0]
})
importances = importances.sort_values(by='Importance', ascending=False)
```

```python
plt.bar(x=importances['Attribute'], height=importances['Importance'], color='#087E8B')
plt.title('Importanacia de las caracterÃ­sticas obtenidas de los coeficientes', size=20)
plt.xticks(rotation='vertical')
plt.show()
```
