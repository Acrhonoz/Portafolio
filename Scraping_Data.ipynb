```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
import time
```

```python
def scraping():
    r = requests.get('https://www.cenace.gob.mx/paginas/publicas/info/demandaregional.aspx')
    soup = BeautifulSoup(r.text, 'lxml')
    zonas = soup.find_all('th')
    zonas = zonas[:9]
    lista_zonas = []
    for zona in zonas:
        lista_zonas.append(zona.text)
    campos = soup.find_all('td',attrs={'class':'num'})
    campos = campos[:27]
    datos = []
    for campo in campos:
        datos.append(float(campo.text.replace(',','').replace('N/A','0').replace('‐','-')))
    vector_datos = np.array(datos)
    matriz_datos = vector_datos.reshape(9,3)
    lista_matricial_datos = matriz_datos.tolist()
    dict_from_list = dict(zip(lista_zonas, lista_matricial_datos)) 
    dataset1 = pd.DataFrame(dict_from_list)
    dataset1[''] = ['Generación_Neta','Intercambio_Neto', 'Demanda_Neta']
    dataset1 = dataset1.set_index('')
    return dataset1
```

```python
# Creamos la automzatización de la consulta
# Esta automatización nos permite scrapear la página durante 4 horas cada 5 minutos, ya que los datos de la pagina
# cambian cada 5 min
diferencia = 0
inicio = time.time()
datasetHis = scraping()
while diferencia < 14400:
    time.sleep(301)
    dataset2 = scraping()
    datasetHis = pd.concat([datasetHis,dataset2])
    fin = time.time()
    diferencia = fin - inicio
```
