# EJERCICIO 1

## 1) Limpiar la columna “Date of Publication” para que sea de tipo numérico. En caso de que exista más de un año ej. “[1860?-70?]” ó “1815, [1816]”, dejar cualquiera de los dos.

```python
# Se Importa el un libro de excel de la biblioteca Brookyn
df = pd.read_excel('/Users/acrhonoz./Downloads/prueba python/fragments/Biblioteca de Brooklyn 1.xlsx')
# Se limpia la columna "Date of Publication"
p = df['Date of Publication'].astype(str).str.extract('(\d+)')
# Se rellenan los NaN con 0, pues son años con esto se homologa el tipo de dato de la columna
p = p.fillna(0)
# Se cambia el tipo dato a "int"
p = p.astype(int)
# Se implementan los cambios a la columna en del dataset original
df['Date of Publication'] = p
```

## 2) Ordenar utilizando la columna “Identifier” de manera ascendente.

```python
# Se crea el dataset con la columna ordenada de forma ascendente 
by_identifier = df.sort_values('Identifier', ascending=True)
# Se muestra que el cambio se haya efectuado
by_identifier
```

## 3) Una vez se tengan los datos ordenados, completar los registros vacíos de la columna “Author” con el autor del registro anterior al que se requiere completar.

```python
# De la columna "Autor" se rellenan los registros con NaN con el nombre del autor de la fila anterior
by_identifier['Author'] = by_identifier['Author'].fillna(method ='ffill')
# Se verifica que el cambio se haya efectuado
by_identifier
```

```python
# Como los libros de excel tienen nombres secuenciales, el proceso realizado anteriormente para 1 libro se 
# puede automatizar de tal forma que lo haga para todos los libros

# Automatización del proceso para los 6 libros de Excel
dataframes = []
a = '1'
for i in range(6):
    df = pd.read_excel(f'/Users/acrhonoz./Downloads/prueba python/fragments/Biblioteca de Brooklyn {a}.xlsx')
    # 1) Limpiar la columna "Date Of Publication"
    p = df['Date of Publication'].astype(str).str.extract('(\d+)')
    p = p.fillna(0)
    p = p.astype(int)
    df['Date of Publication'] = p
    # 2) Ordenar utilizando la columna “Identifier” de manera ascendente.
    by_identifier = df.sort_values('Identifier', ascending=True)
    # 3) Una vez se tengan los datos ordenados, completar los registros vacíos de la columna “Author” con el autor del registro anterior al que se requiere completar.
    by_identifier['Author'] = by_identifier['Author'].fillna(method ='ffill')
    dataframes.append(by_identifier)
    a = '0' + a
```

```python
# Aunque no se especifica, me tomé el atrevimiento que compilar todos datasets en un solo, cada uno limpio y 
# procesado
# Unión de los 6 Dataframes en 1 solo.
data_frame_completo = pd.concat([dataframes[0],dataframes[1],dataframes[2],dataframes[3],dataframes[4],dataframes[5]], axis=0)
```

```python
# Al compilar se puede revolver el orden ascendente por el campo "Identifier", por lo que 
# se reordena y se vuelven a llenar los NaN con el autor del renglón anterior
data_frame_completo = data_frame_completo.sort_values('Identifier', ascending=True)
data_frame_completo['Author'] = data_frame_completo['Author'].fillna(method ='ffill')
# Se muestra el dataset completo
data_frame_completo
```

## Crear y entregar un archivo CSV o XLSX con las fechas donde existe cambio de horario en México. Las fechas deberán correr desde 1996 hasta el año 2100. Al año se tienen dos fechas con cambio de horario: el primer domingo de cada abril y el último domingo de octubre.

```python
# Haciendo uso de la librería "calendar" se obtiene el primer domingo de Abril y el último domingo de Octubre
# de esta forma se localizan los principios y finales del horario de verano
# se repite el proceso, iniciando en 1996 y finalizando en 2100
# y se Imprimen las fechas en las habrá inicio y fin de horario de verano

for i in range(1996,2101):
    abril = calendar.month(i,4)
    print('Inicio de horario de verano: ' + abril[56] + '/' + 'Abril' + '/' + str(i))
    octubre = calendar.month(i,10)
    octubre_s = octubre.split("\n")
    octubre_a = np.array(octubre_s[2:])
    if len(octubre_a[4]) != 20:
        print('Termina Horario de verano: ' + octubre_a[3][18] + octubre_a[3][19] +'/'+ 'Octubre' +'/'+ str(i))
    else:
        print('Termina Horario de verano: ' + octubre_a[4][18] + octubre_a[4][19] +'/'+ 'Octubre' +'/'+ str(i))    
    print()
```

```python
# Creo una subrutina que me almacene cada inicio y cada termino en 2 listas, una para tipo de dato
# con ello creo un diccionario que tenga 2 campos "Inicio" y 'Termino'
# Transformo ese diccionario a un dataframe y lo exporto a un libro de excel
# las fechas quedan con el formaro "aaaammd" para el inicio y "aaaammdd" para el termino
lista_inicio = []
lista_termino = []
dic = {'Inicio': '', 'Termino': ''} 
for i in range(1996,2101):
    abril = calendar.month(i,4)
    lista_inicio.append(int(str(i) + '04' + abril[56]))    
    octubre = calendar.month(i,10)
    octubre_s = octubre.split("\n")
    octubre_a = np.array(octubre_s[2:])
    if len(octubre_a[4]) != 20:
        lista_termino.append(int(str(i) + '10' + octubre_a[3][18] + octubre_a[3][19]))
    else:
        lista_termino.append(int(str(i) + '10' + octubre_a[4][18] + octubre_a[4][19]))
dic['Inicio'] = lista_inicio
dic['Termino'] = lista_termino
df_Ho_ver = pd.DataFrame(dic)
df_Ho_ver.to_excel('/Users/acrhonoz./Desktop/Horario_Verano.xlsx')
```
